### Import Libs

```python
import pandas as pd
import spacy
import nltk
nltk.download('stopwords')
from api_model import nlextract

pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', None)
```

```python
extractor = nlextract.NLExtractor()
```

## Read Excel

```python
caminho = '/opt/dna/find-keywords/datalake/Conversas Sodexo.xlsx'
df = pd.read_excel(caminho, engine='openpyxl')
df.info()
```

```python
df.head()
```

```python
df = df.sample(n=10000)
```

```python
df['Message'].count()
```

```python
coluna_texto = 'Message'
df[f'{coluna_texto}_clean'] = df[coluna_texto].apply(extractor.udf_clean_text) ## limpa toda a acentuação, caracteres especiais,
                                                                                  # pontuações do texto.
```

```python
df[[f'{coluna_texto}_clean', 'Message']].head(35)
```

```python
## 4 digitos
patters = '{4}\s|\d{4}|(?:\.|,|[0-9]{4})*|\d{4}'
## 10 digitos
#patters = '{10}'
```

```python
## extrair 4 ou 10 digitos
df[f'{coluna_texto}_extract_digits'] = df[f'{coluna_texto}_clean'].apply(lambda x: extractor.udf_extract_digits(x, patters))
```
